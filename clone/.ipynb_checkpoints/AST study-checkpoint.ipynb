```python
trees = source.set_index('id',drop=False).loc[train_ids]
```

```python
from prepare_data import get_sequences

def trans_to_sequences(ast):
    sequence = []
    get_sequences(ast, sequence)
    return sequence
```

```python
corpus = trees['parsed code'].apply(trans_to_sequences) # type = Series(list(str))
```

```python
str_corpus = [' '.join(c) for c in corpus]  # type = list(str)
```

```python
from gensim.models.word2vec import Word2Vec
w2v = Word2Vec(corpus, size=128, workers=16, sg=1, max_final_vocab=3000)
```

```python
word2vec = w2v.wv
vocab = word2vec.vocab  # type = dict
max_token = word2vec.vectors.shape[0]
# word2vec['i']  'i' 
```

```python
del train_ids, trees, corpus, str_corpus  # This only serve for computing w2v
```

```python
from prepare_data import get_blocks

def tree_to_index(node):
    token = node.token
    result = [vocab[token].index if token in vocab else max_token]
    children = node.children
    for child in children:
        result.append(tree_to_index(child))
    return result

def trans2seq(r):
    blocks = []
    get_blocks(r, blocks)
    tree = []
    for b in blocks:
        btree = tree_to_index(b)
        tree.append(btree)
    return tree
```

```python
blocks = pd.DataFrame(source, copy=True)
blocks['code'] = blocks['code'].apply(trans2seq)
if 'label' in blocks.columns:
    blocks.drop('label', axis=1, inplace=True)
```
